#### Nombres y créditos de clientes de los compradores de Pink Cake (Enlugar de majadas) ordenado por crédito descendente y nombre ascendente

```python
%%sql SELECT cliente.nombre, credito FROM cliente
INNER JOIN compra
ON cliente.id = compra.cliente_id
INNER JOIN tienda
ON tienda_id = tienda.id
WHERE tienda.nombre = 'Pink Cake'
ORDER BY credito DESC, cliente.nombre ASC;
```

#### Agregamos credito en dolares a un cambio de 7.8

```python
%%sql SELECT cliente.nombre, credito, (credito * 7.8) AS credito_dolares FROM cliente
INNER JOIN compra
ON cliente.id = compra.cliente_id
INNER JOIN tienda
ON tienda_id = tienda.id
WHERE tienda.nombre = 'Pink Cake'
ORDER BY credito DESC, cliente.nombre ASC;
```

#### Parejas de clientes con el mismo crédito que no sean ellos mismos y que no haya parejas duplicadas

```python
%%sql SELECT DISTINCT cliente.nombre AS cliente1, cust.nombre AS cliente2, cliente.credito FROM cliente
INNER JOIN (SELECT * FROM cliente) AS cust
ON cliente.credito = cust.credito
WHERE cliente.nombre != cust.nombre;
```

#### Clientes que han comprado en Pink Cake ( En lugar de majadas) y no en El encanto (En lugar de VH)

```python
%%sql SELECT pink.nombre FROM (
	SELECT cliente.nombre, tienda.nombre FROM cliente
	INNER JOIN compra
	ON cliente.id = compra.cliente_id
	INNER JOIN tienda
	ON (tienda.id = compra.tienda_id)
	WHERE tienda.nombre = 'Pink Cake'
) AS pink
INNER JOIN (
	SELECT cliente.nombre, tienda.nombre FROM cliente
	INNER JOIN compra
	ON cliente.id = compra.cliente_id
	INNER JOIN tienda
	ON (tienda.id = compra.tienda_id)
	WHERE tienda.nombre != 'El encanto'
) AS encanto
ON encanto.nombre = pink.nombre;
```

#### Todas las tiendas que tengan otras tiendas en la misma dirección

```python
%%sql SELECT DISTINCT tienda.nombre AS tienda1, tienda2.nombre AS tienda2, tienda.direccion FROM tienda
INNER JOIN (SELECT * FROM tienda) AS tienda2
ON tienda.direccion = tienda2.direccion
WHERE tienda.nombre != tienda2.nombre;
```

#### Clientes con los créditos más altos (sin utilizar MAX()

```python
%%sql SELECT * FROM cliente
ORDER BY credito DESC;
```

#### Clientes cuyo crédito en dólares sea mayor a 1 (Tasa de cambio 7.8)

```python
%%sql SELECT * FROM cliente
WHERE (credito * 7.8) > 1
ORDER BY credito DESC;
```

#### Personas que no han hecho ninguna compra

```python
%%sql SELECT cliente.nombre FROM cliente
LEFT JOIN compra
ON compra.cliente_id = cliente.id
WHERE compra.id IS NULL;
```

#### Diferencia entre el promedio de crédito de los clientes de Pink cake (En lugar de Majadas) y de el encanto (En lugar de VH)

```python
%%sql SELECT (avg_encanto - avg_pink) AS diferencia_avg FROM (
	SELECT avg(credito) AS avg_encanto FROM cliente
	INNER JOIN compra
	ON compra.cliente_id = cliente.id
	INNER JOIN tienda
	ON compra.tienda_id = tienda.id
	WHERE tienda.nombre = 'El encanto'
) JOIN (
	SELECT avg(credito) AS avg_pink FROM cliente
	INNER JOIN compra
	ON compra.cliente_id = cliente.id
	INNER JOIN tienda
	ON compra.tienda_id = tienda.id
	WHERE tienda.nombre = 'Pink Cake'
);
```

#### Número de clientes de cada tienda, de las que tienen compras, ordenado por nombre de tienda.

```python
%%sql SELECT tienda.nombre, count(*) AS numero_clientes FROM compra
INNER JOIN tienda
ON tienda.id = compra.tienda_id
GROUP BY tienda_id;
```
