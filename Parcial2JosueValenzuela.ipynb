##### ¿Qué es una base de datos?

Una base de datos es una recolección de datos con un esquema definido y ordenado, facilitando así su acceso tanto para lectura como para escritura de datos.

#### Proceso de Normalización

El objetivo de la normalización de datos es disminuir la redundancia en el almacenamiento de datos. Así lograr una mayor eficiencia en espacio evitando almacenar registros duplicados.

#### ¿Cuál es el objetivo principal de la Inteligencia de Negocios? ¿Cómo se relaciona este concepto con el concepto de DataWareHouse?

La inteligencia de negocios es básicamente tomar decisiones de una manera crítica y objetiva (en un negocio) basados en datos reales y tangibles.

Los datos son muy útiles para poder tomar decisiones de negocios basadas en la realidad. Los datos no siempre son fáciles de entender a menos que se posea un conocimiento técnico sobre el tema. El data warehouse es introducido para optimizar el proceso de consulta de datos con el objetivo de tener una visualización grafica y sencilla de comprender para personas que quizás no cuenten con los conocimientos técnicos que de otra forma necesitarían para comprender los datos.

#### Nombres y créditos de clientes de los compradores de Pink Cake (Enlugar de majadas) ordenado por crédito descendente y nombre ascendente

```python
%%sql SELECT cliente.nombre, credito FROM cliente
INNER JOIN compra
ON cliente.id = compra.cliente_id
INNER JOIN tienda
ON tienda_id = tienda.id
WHERE tienda.nombre = 'Pink Cake'
ORDER BY credito DESC, cliente.nombre ASC;
```

#### Agregamos credito en dolares a un cambio de 7.8

```python
%%sql SELECT cliente.nombre, credito, (credito * 7.8) AS credito_dolares FROM cliente
INNER JOIN compra
ON cliente.id = compra.cliente_id
INNER JOIN tienda
ON tienda_id = tienda.id
WHERE tienda.nombre = 'Pink Cake'
ORDER BY credito DESC, cliente.nombre ASC;
```

#### Parejas de clientes con el mismo crédito que no sean ellos mismos y que no haya parejas duplicadas

```python
%%sql SELECT DISTINCT cliente.nombre AS cliente1, cust.nombre AS cliente2, cliente.credito FROM cliente
INNER JOIN (SELECT * FROM cliente) AS cust
ON cliente.credito = cust.credito
WHERE cliente.nombre != cust.nombre;
```

#### Clientes que han comprado en Pink Cake ( En lugar de majadas) y no en El encanto (En lugar de VH)

```python
%%sql SELECT pink.nombre FROM (
	SELECT cliente.nombre, tienda.nombre FROM cliente
	INNER JOIN compra
	ON cliente.id = compra.cliente_id
	INNER JOIN tienda
	ON (tienda.id = compra.tienda_id)
	WHERE tienda.nombre = 'Pink Cake'
) AS pink
INNER JOIN (
	SELECT cliente.nombre, tienda.nombre FROM cliente
	INNER JOIN compra
	ON cliente.id = compra.cliente_id
	INNER JOIN tienda
	ON (tienda.id = compra.tienda_id)
	WHERE tienda.nombre != 'El encanto'
) AS encanto
ON encanto.nombre = pink.nombre;
```

#### Todas las tiendas que tengan otras tiendas en la misma dirección

```python
%%sql SELECT DISTINCT tienda.nombre AS tienda1, tienda2.nombre AS tienda2, tienda.direccion FROM tienda
INNER JOIN (SELECT * FROM tienda) AS tienda2
ON tienda.direccion = tienda2.direccion
WHERE tienda.nombre != tienda2.nombre;
```

#### Diferencia entre el promedio de crédito de los clientes de Pink cake (En lugar de Majadas) y de el encanto (En lugar de VH)

```python
%%sql SELECT (avg_encanto - avg_pink) AS diferencia_avg FROM (
	SELECT avg(credito) AS avg_encanto FROM cliente
	INNER JOIN compra
	ON compra.cliente_id = cliente.id
	INNER JOIN tienda
	ON compra.tienda_id = tienda.id
	WHERE tienda.nombre = 'El encanto'
) JOIN (
	SELECT avg(credito) AS avg_pink FROM cliente
	INNER JOIN compra
	ON compra.cliente_id = cliente.id
	INNER JOIN tienda
	ON compra.tienda_id = tienda.id
	WHERE tienda.nombre = 'Pink Cake'
);
```

#### Número de clientes de cada tienda, de las que tienen compras, ordenado por nombre de tienda.

```python
%%sql SELECT tienda.nombre, count(*) AS numero_clientes FROM compra
INNER JOIN tienda
ON tienda.id = compra.tienda_id
GROUP BY tienda_id;
```

#### Cree un trigger que considere apropiado para este esquema. Justifique su elección (5)

Trigger resta el total de una compra del credito del cliente al momento en el que se inserta una compra, de esta manera mantener el credito del cliente actualizado.

```python
%%sql CREATE TRIGGER change_credit AFTER INSERT ON compra
BEGIN
	UPDATE cliente SET credito = (credito - new.total) WHERE cliente.id = new.cliente_id;
END;
```

#### Genere un reporte (vista) que muestre información de ventas a ser utilizado por los gerentes de la tienda. 

Vista genera reporte del totol de ventas por mes (de cada anio) de las tiendas. Esto para faculitar el analisis de las ventas por mes.

```python
%%sql CREATE VIEW ventas_por_mes AS
SELECT tienda.nombre AS nombre_tienda, sum(total) as total_ventas, strftime('%m', DATE(fecha)) as mes, strftime('%Y', DATE(fecha)) as anio FROM compra
INNER JOIN tienda
ON compra.tienda_id = tienda.id
GROUP BY mes, anio, tienda.nombre
ORDER BY anio DESC;
```

```python
%sql SELECT * FROM ventas_por_mes
```
